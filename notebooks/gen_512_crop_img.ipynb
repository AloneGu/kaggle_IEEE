```python
import glob
import os
import numpy as np
from PIL import Image
import skimage
import skimage.io
import scipy.misc
import imageio
import warnings; warnings.simplefilter('ignore')

cls = os.listdir('../input/train')
print(cls)
for c in cls:
    p = os.path.join('../input/crop_train',c)
    if not os.path.exists(p):
        os.makedirs(p)
```

```python
import numpy as np
train_dir = '../input/train'
output_dir = '../input/crop_train'
train_files = sorted(glob.glob(train_dir+'/*/*'))
CROP_LEN = 512

def random_crop(im_array):
    # crop
    x_range = im_array.shape[0] - CROP_LEN
    y_range = im_array.shape[1] - CROP_LEN
    # print(x_range,y_range)
    a = np.random.randint(x_range)
    b = a + CROP_LEN
    c = np.random.randint(y_range)
    d = c + CROP_LEN
    new_im_array = im_array[a:b,c:d,:]
    return new_im_array

def random_manip(img_p):
    img_name = os.path.basename(img_p)
    img_name = os.path.splitext(img_name)[0]
    img_cnt = 1
    cls = img_p.split('/')[-2]
    im_array = np.array(Image.open((img_p)), dtype="uint8")
    
    # no manip
    for i in range(8):
        tmp_img = random_crop(im_array)
        output_p = os.path.join(output_dir,cls,"{}_{}_unalt.jpg".format(img_name,img_cnt))
        img_cnt += 1
        imageio.imwrite(output_p,tmp_img,quality=100)
        
    tmp_img = skimage.exposure.adjust_gamma(random_crop(im_array), gamma=0.8)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=100)
    
    tmp_img = skimage.exposure.adjust_gamma(random_crop(im_array), gamma=1.2)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=100)
    
    tmp_img = random_crop(im_array)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=70)
    
    tmp_img = random_crop(im_array)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=90)
    
    new_img = scipy.misc.imresize(random_crop(im_array), 2.0, interp='bicubic')
    tmp_img = random_crop(new_img)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=100)
    
    new_img = scipy.misc.imresize(random_crop(im_array), 1.5, interp='bicubic')
    tmp_img = random_crop(new_img)
    output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
    img_cnt += 1
    imageio.imwrite(output_p,tmp_img,quality=100)
    
    try:
        new_img = scipy.misc.imresize(im_array, 0.5, interp='bicubic')
        tmp_img = random_crop(new_img)
        output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
        img_cnt += 1
        imageio.imwrite(output_p,tmp_img,quality=100)
    except:
        print('small img',img_p)
        pass
    
    try:
        new_img = scipy.misc.imresize(im_array, 0.8, interp='bicubic')
        tmp_img = random_crop(new_img)
        output_p = os.path.join(output_dir,cls,"{}_{}_manip.jpg".format(img_name,img_cnt))
        img_cnt += 1
        imageio.imwrite(output_p,tmp_img,quality=100)
    except:
        print('small img',img_p)
        pass

process_cnt = 0
for p in train_files:
    if process_cnt < 1000:
        process_cnt += 1
        continue # have done 1000 images
    else:
        random_manip(p)
        if process_cnt % 100 == 0:
            print(process_cnt,p)
        process_cnt += 1
```
